```python
#importing pandas library

import pandas as pd

#Creating a data frame df for data from excel
df = pd.read_csv('Fiverr.csv')
df.head(3)
```

```python
#creating pivot table
pivot = df.pivot_table(values = 'line value', index = ['stripe','mgmt stripe','PLANNER_CODE','END_ITEM_NAME','END_ITEM_DESCRIPTION','ITEM_NAME','ITEM_DESCRIPTION'],columns = 'DEMAND_DATE', aggfunc = 'sum')
```

```python
#showing the pivot table
pivot.head()
```

```python
#saving pivot table to excel sheet
pivot.to_excel('PIVOT.xlsx')
```
